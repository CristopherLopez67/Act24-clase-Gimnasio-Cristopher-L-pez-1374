```python
# Definición de la estructura para Gimnasio
class Gimnasio:
    def __init__(self, nombre, capacidad, tipoMembresia, equipamiento):
        # Atributos
        self.nombre = nombre
        self.capacidad = capacidad
        self.tipoMembresia = tipoMembresia
        self.equipamiento = equipamiento

    # Funciones
    def registrarEntrada(self):
        return f"En {self.nombre}, se ha registrado la entrada de un usuario."

    def asignarRutina(self):
        return f"El gimnasio {self.nombre} ha asignado una rutina personalizada."

    def medirProgreso(self):
        return f"Se está midiendo el progreso del usuario en {self.nombre}."

# --- Creación de los objetos ---
gimnasio_norte = Gimnasio("FitZone", 100, "Premium", "Pesas y Cardio")
gimnasio_sur = Gimnasio("PowerGym", 50, "Básica", "Máquinas de fuerza")

# --- Ejecución de las funciones y mensajes ---

# Datos del gimnasio 1
print(f"--- Ficha de {gimnasio_norte.nombre} ---")
print(f"Capacidad: {gimnasio_norte.capacidad} | Membresía: {gimnasio_norte.tipoMembresia}")
print(f"Equipamiento: {gimnasio_norte.equipamiento}")
print(gimnasio_norte.registrarEntrada())
print(gimnasio_norte.asignarRutina())
print(gimnasio_norte.medirProgreso())

print("\n")

# Datos del gimnasio 2
print(f"--- Ficha de {gimnasio_sur.nombre} ---")
print(f"Capacidad: {gimnasio_sur.capacidad} | Membresía: {gimnasio_sur.tipoMembresia}")
print(f"Equipamiento: {gimnasio_sur.equipamiento}")
print(gimnasio_sur.registrarEntrada())
print(gimnasio_sur.asignarRutina())
print(gimnasio_sur.medirProgreso())
```
